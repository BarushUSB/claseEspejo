```python
#Carga de librerias
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
```

```python
#Cargamos las combinaciones de la compuerta AND
training_data=np.array([[0,0],[0,1],[1,0],[1,1]], "float32")
```

```python
#Cargamos las deseadas para cada combinación de AND
target_data=np.array([[0],[0],[0],[1]], "float32")
```

```python
#Creacion del modelo
model = Sequential()
model.add(Dense(3, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
#Compilación y ejecución del modelo
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['binary_accuracy'])
model.fit(training_data, target_data, epochs=100)
resultados=model.evaluate(training_data, target_data)
print("\n%s: %.2f%%" % (model.metrics_names[1], resultados[1]*100))
```
